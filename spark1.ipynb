```python
from pyspark import SparkContext
sc = SparkContext.getOrCreate()
```

```python
data=[1,2,3,4,5]
rdd = sc.parallelize(data)
```

```python
#Transformation
rdd_squared = rdd.map(lambda x: x ** 2)
```

```python
# Action
result = rdd_squared.reduce(lambda x, y: x + y)
# Print the result
print(result)
```

```python
###UNION
union_inp = sc.parallelize([2,4,5,6,7,8,9])
union_rdd_1 = union_inp.filter(lambda x: x % 2 == 0)
union_rdd_2 = union_inp.filter(lambda x: x % 3 == 0)
print(union_rdd_1.union(union_rdd_2).collect())
```

Q4. Perform the following operations using RDDs:

1. Create an RDD with the following list of course names:
["Python", "Spark", "Hadoop", "Python", "Hive", "Spark", "Scala", "Hadoop", "SQL"]
2.Print all course names.

3. Count the total number of courses.

4. Find the distinct course names.

5. Convert all course names to uppercase.

6. Sort the course names alphabetically.

7. Print course names containing the letter 'a'.

```python
from pyspark import SparkContext
sc = SparkContext.getOrCreate()
course_rdd=sc.parallelize(["Python", "Spark", "Hadoop", "Python", "Hive", "Spark", "Scala", "Hadoop", "SQL"])


#2
print(course_rdd.count())

#3
print(course_rdd.distinct().collect())
#4
upper_rdd=course_rdd.map(lambda x: x.upper())
print(upper_rdd.collect())
#5
sorted_rdd=course_rdd.sortBy(lambda x: x)
print(sorted_rdd.collect())

#6
a_rdd=course_rdd.filter(lambda x: 'a' in x)
print(a_rdd.collect())
```

Q5. Perform the following operations using RDDs:

1. Create an RDD with the following sentences:
["Spark makes big data processing fast",
 "Python is widely used in data science",
 "Big data tools include Spark and Hadoop",
 "Python and Spark work well together"]
 2. Print all sentences.
 3. Count the number of sentences.
 4. Split the sentences into words.
 5. Print words containing the letter 'a'.
 6. Convert all words to uppercase.
 7. Count the total number of words.

```python
sentence_rdd=sc.parallelize( ["Spark makes big data processing fast", "Python is widely used in data science", "Big data tools include Spark and Hadoop", "Python and Spark work well together"])

#2
print(sentence_rdd.collect())

#3
print(sentence_rdd.count())
#4
words_rdd=sentence_rdd.flatMap(lambda x: x.split())
print(words_rdd.collect())
#5
a_words_rdd=words_rdd.filter(lambda x: 'a' in x)
print(a_words_rdd.collect())
# sort Alphabetically
sorted_words_rdd=words_rdd.sortBy(lambda x: x)
print(sorted_words_rdd.collect())
#6
upper_words_rdd=words_rdd.map(lambda x: x.upper())
print(upper_words_rdd.collect())
#7
print(words_rdd.count())
```

## WORD COUNT USING MAP-REDUCE


```python
from pyspark import SparkContext
sc = SparkContext.getOrCreate()

#read the input
data=["nagpur nagpur delhi raipur delhi nagpur"]
text_file=sc.parallelize(data)
#read dtaa frim text file
words=text_file.flatMap(lambda line: line.split(" "))

#word count
wcount=words.map(lambda word: (word,1))
print(wcount.collect())

#reduce
rcount=wcount.reduceByKey(lambda x,y: x+y)
print(rcount.collect())

#save
rcount.saveAsTextFile("Op.txt")

```
