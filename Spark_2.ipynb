#Q1. How would you combine transaction logs from two different days into a single dataset?
Scenario:You receive daily transaction logs and want to analyze them together.

```python
from pyspark import SparkContext
sc = SparkContext.getOrCreate()
day1 = sc.parallelize(["TXN1", "TXN2", "TXN3"])
day2 = sc.parallelize(["TXN4", "TXN5"])

all_txns = day1.union(day2)

print(all_txns.collect())
```

A company wants to identify users who registered on both mobile and web apps.

```python
mobile_users = sc.parallelize(["u1","u2","u3","u4"])
web_users = sc.parallelize(["u3","u4","u5"])
common_users = mobile_users.intersection(web_users)
print(common_users.collect())
```

products sold in both online and offline stores?

```python
online_products = sc.parallelize(["p1","p2","p3"])
offline_products = sc.parallelize(["p2","p3","p4"])
common_products = online_products.intersection(offline_products)
```

You want users who signed up but never bought anything.

```python
registered_users = sc.parallelize(["u1","u2","u3","u4"])
buyers = sc.parallelize(["u2","u3"])
non_buyers = registered_users.subtract(buyers)
print(non_buyers.collect())
```

You want to analyze every product in every region.

```python
products = sc.parallelize(["Laptop","Phone"])
regions = sc.parallelize(["US","EU"])
product_region = products.cartesian(regions)
print(product_region.collect())
```

Q1. Dataframe Operations
1. Create Create an RDD from the given data.
2. Convert the RDD into a DataFrame with the schema: name (string), roll_no (integer),marks (integer)
3. Display the DataFrame.
4. Calculate the total sum of marks of all students using DataFrame operations.
5. Display the result in the format

```python
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName("StudentMarks").getOrCreate()
sc = spark.sparkContext
data = [
    ("Alice", 101, 85),
    ("Bob", 102, 90),
    ("Cathy", 103, 78),
    ("David", 104, 92)
]
#Create RDD
rdd = sc.parallelize(data)

#Convert RDD to Dataframe
df = spark.createDataFrame(rdd, ["name", "roll_no", "marks"])
#Display
df. show()

# Find total sum of marks
from pyspark.sql.functions import sum
total_marks = df.select(sum("marks")).collect()
total_marks1=total_marks[0][0]
print("Total Marks =", total_marks)
print("Total Marks =", total_marks1)
```

Q1. E-commerce Order Analysis: You work as a data engineer in an e-commerce company. The company stores order data in a Spark DataFrame with the following schema:

| order_id | customer_id | product | category | price | quantity | order_date |

Questions
1. Calculate total revenue for each category.
2. Find the top 5 customers who spent the most money.
3. Identify orders where quantity > 10.
4. Find daily total sales.
5. Find the most popular product (highest quantity sold).

```python
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName("EcommerceAnalysis").getOrCreate()
data = [
(1,101,"Laptop","Electronics",50000,1,"2024-01-01"),
(2,102,"Phone","Electronics",20000,2,"2024-01-02"),
(3,101,"Shoes","Fashion",3000,3,"2024-01-02"),
(4,103,"Watch","Fashion",7000,1,"2024-01-03"),
(5,104,"Laptop","Electronics",50000,1,"2024-01-03")
]

columns = ["order_id","customer_id","product","category","price","quantity","order_date"]

df = spark.createDataFrame(data, columns)
```

```python
###Calculate Total Revenue for Each Category
##Revenue = price × quantity
from pyspark.sql.functions import col, sum
df.withColumn("revenue", col("price") * col("quantity")).groupBy("category") \
.agg(sum("revenue").alias("total_revenue")).show()
```

```python
##Top 5 Customers Who Spent the Most
df.withColumn("revenue", col("price") * col("quantity")).groupBy("customer_id") \
.agg(sum("revenue").alias("total_spent"))\
.orderBy(col("total_spent").desc()).limit(5).show()
```

```python
###Identify Orders Where Quantity > 10
df.filter(col("quantity") > 10).show()
```

```python
###Find Daily Total Sales
df.withColumn("revenue", col("price") * col("quantity")) \
  .groupBy("order_date") \
  .agg(sum("revenue").alias("daily_sales")) \
  .show()
```

```python
##Most Popular Product (Highest Quantity Sold)
from pyspark.sql.functions import sum

df.groupBy("product") \
  .agg(sum("quantity").alias("total_quantity")) \
  .orderBy(col("total_quantity").desc()) \
  .limit(1) \
  .show()
```

Q2.  Student Performance Dataset
Dataset: | student_id | subject | marks | exam_date |

Questions:
1. Find average marks per subject.
2. Find top scoring student in each subject.
3. List students scoring below 40.
4. Calculate total marks per student.

```python
from pyspark.sql import SparkSession

spark = SparkSession.builder.appName("StudentPerformance").getOrCreate()

data = [
(1,"Math",85,"2024-03-01"),
(2,"Math",78,"2024-03-01"),
(3,"Math",92,"2024-03-01"),
(1,"Science",75,"2024-03-02"),
(2,"Science",35,"2024-03-02"),
(3,"Science",88,"2024-03-02"),
(4,"Math",39,"2024-03-01")
]

columns = ["student_id","subject","marks","exam_date"]

df = spark.createDataFrame(data, columns)
```

```python
###Find Average Marks per Subject
from pyspark.sql.functions import avg

df.groupBy("subject") \
  .agg(avg("marks").alias("average_marks")) \
  .show()
```

```python
#Find Top Scoring Student in Each Subject
from pyspark.sql.window import Window
from pyspark.sql.functions import rank, col

## Splits the dataset into independent groups based on subject.
## Each subject (e.g., Math, Science) is processed separately.

windowSpec = Window.partitionBy("subject").orderBy(col("marks").desc())

#Add a new column called rank , by Applying the ranking function over each subject partition.
df.withColumn("rank", rank().over(windowSpec)).show()

df.withColumn("rank", rank().over(windowSpec)) \
  .filter(col("rank") == 1) \
  .select("student_id","subject","marks") \
  .show()
```

```python
##List Students Scoring Below 40
df.filter(col("marks") < 40).show()
```

```python
##Calculate Total Marks per Student
from pyspark.sql.functions import sum

df.groupBy("student_id") \
  .agg(sum("marks").alias("total_marks")) \
  .show()
```

Linear Regression with one feature and one class

```python
from pyspark.sql import SparkSession
from pyspark.ml.regression import LinearRegression
from pyspark.ml.feature import VectorAssembler
from pyspark.ml.evaluation import RegressionEvaluator
# Initialize Spark session
spark = SparkSession.builder.appName("LinearRegressionExample").getOrCreate()
```

```python
data_df = spark.createDataFrame([(1.0, 2.2), (2.0, 3.9), (3.0, 5.8), (4.0, 7.1)], ["feature", "label"])
```


VectorAssembler to assemble your features into a single column:


```python
assembler = VectorAssembler(inputCols=["feature"], outputCol="features")
data_df = assembler.transform(data_df)
```

```python
# Split the dataset in train and Test
#Seed fixes the randomness and Guarantees Reproducibility → same split every time you run it
train_data, test_data = data_df.randomSplit([0.8, 0.2], seed=1234)
```

```python
lr = LinearRegression(featuresCol="features", labelCol="label")
lr_model = lr.fit(train_data)
```

```python
trainingSummary = lr_model.summary
print("RMSE: %f" % trainingSummary.rootMeanSquaredError)
print("r2: %f" % trainingSummary.r2)
```

```python
# Optionally, you can also print the coefficients and intercept of the linear model
print("Coefficients:", lr_model.coefficients)
print("Intercept:", lr_model.intercept)
```

Logistic Regression

```python
from pyspark.sql import SparkSession
from pyspark.ml.feature import VectorAssembler
from pyspark.ml.classification import LogisticRegression
from pyspark.ml.evaluation import BinaryClassificationEvaluator

# 1. Create Spark session
spark = SparkSession.builder.appName("LogisticRegressionExample").getOrCreate()
# 2. Sample Data (features + label)
data = [
    (1, 2.0, 1.0, 0),
    (2, 3.0, 2.0, 0),
    (3, 4.0, 3.0, 1),
    (4, 5.0, 4.0, 1),
    (5, 6.0, 5.0, 1)
]

columns = ["id", "feature1", "feature2", "label"]
df = spark.createDataFrame(data, columns)

# 3. Convert features into vector
assembler = VectorAssembler(inputCols=["feature1", "feature2"],outputCol="features")
df = assembler.transform(df)

# 4. Train-Test Split
train_data, test_data = df.randomSplit([0.8, 0.2], seed=1234)

# 5. Initialize Logistic Regression model
lr = LogisticRegression(featuresCol="features",labelCol="label",maxIter=10)

# 6. Train model
model = lr.fit(train_data)

# 7. Make predictions
predictions = model.transform(test_data)

# 8. Show results
predictions.select("id", "features", "label", "prediction", "probability").show()

# 9. Evaluate model
evaluator = BinaryClassificationEvaluator(labelCol="label",metricName="areaUnderROC")

auc = evaluator.evaluate(predictions)
print(f"AUC: {auc}")
```
