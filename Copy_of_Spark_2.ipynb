#Q1. How would you combine transaction logs from two different days into a single dataset?
Scenario:You receive daily transaction logs and want to analyze them together.

```python
from pyspark import SparkContext
sc = SparkContext.getOrCreate()
day1 = sc.parallelize(["TXN1", "TXN2", "TXN3"])
day2 = sc.parallelize(["TXN4", "TXN5"])
day1.union(day2).collect()

## Write code


```

A company wants to identify users who registered on both mobile and web apps.

```python
mobile_users = sc.parallelize(["u1","u2","u3","u4"])
web_users = sc.parallelize(["u3","u4","u5"])

#Write
mobile_users.intersection(web_users).collect()
```

products sold in both online and offline stores?

```python
online_products = sc.parallelize(["p1","p2","p3"])
offline_products = sc.parallelize(["p2","p3","p4"])

#Write
online_products.intersection(offline_products).collect()
```

You want users who signed up but never bought anything.

```python
registered_users = sc.parallelize(["u1","u2","u3","u4"])
buyers = sc.parallelize(["u2","u3"])

# Write
registered_users.subtract(buyers).collect()
```

You want to analyze every product in every region.

```python
products = sc.parallelize(["Laptop","Phone"])
regions = sc.parallelize(["US","EU"])

#Write
products.cartesian(regions).collect()
```

Q1. E-commerce Order Analysis: You work as a data engineer in an e-commerce company. The company stores order data in a Spark DataFrame with the following schema:

| order_id | customer_id | product | category | price | quantity | order_date |

Questions
1. Calculate total revenue for each category.
2. Find the top 5 customers who spent the most money.
3. Identify orders where quantity > 10.
4. Find daily total sales.
5. Find the most popular product (highest quantity sold).

```python
from pyspark.sql import SparkSession

spark = SparkSession.builder.appName("EcommerceAnalysis").getOrCreate()

data = [
(1,101,"Laptop","Electronics",50000,1,"2024-01-01"),
(2,102,"Phone","Electronics",20000,2,"2024-01-02"),
(3,101,"Shoes","Fashion",3000,3,"2024-01-02"),
(4,103,"Watch","Fashion",7000,1,"2024-01-03"),
(5,104,"Laptop","Electronics",50000,1,"2024-01-03")
]

columns = ["order_id","customer_id","product","category","price","quantity","order_date"]

df = spark.createDataFrame(data, columns)
```
